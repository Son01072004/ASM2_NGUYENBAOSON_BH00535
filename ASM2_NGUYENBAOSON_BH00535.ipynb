```python
# Replace NaN values in 'Discount Applied' column with 'No discount'
df1['Discount Applied'] = df1['Discount Applied'].fillna('No discount')

# Replace NaN values in 'Promotion Used' column with 'No promotion'
df1['Promotion Used'] = df1['Promotion Used'].fillna('No promotion')
df1
```

```python
import matplotlib.pyplot as plt

# Sample data
payment_methods = ['Credit Card', 'Debit Card', 'Cash']
payment_counts = [125, 80, 45]

# Create the pie chart
plt.pie(payment_counts, labels=payment_methods, autopct='%1.1f%%')
plt.title('Payment Method Analysis')

# Display the chart
plt.axis('equal')
plt.show()
```

```python
product_quantity = df1.groupby('Product Name')['Quantity Sold'].sum()

# Create the column chart
product_quantity.plot(kind='bar')

# Set chart title and labels
plt.title('Quantity Sold by Product')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')

# Display the chart
plt.show()
```

```python
product_revenue = df1.groupby('Product Name')['Sale Amount'].sum()

# Create the pie chart
product_revenue.plot(kind='pie', autopct='%1.1f%%')

# Set chart title
plt.title('Revenue by Product')

# Display the chart
plt.show()
```

```python
# Assuming your DataFrame is named df
df1['Transaction Date'] = pd.to_datetime(df1['Transaction Date'])
df1['Transaction Month'] = df1['Transaction Date'].dt.month

transaction_counts = df1['Transaction Month'].value_counts().sort_index()

# Create the line chart
transaction_counts.plot(kind='line', marker='o')

# Set chart title and labels
plt.title('Number of Transactions by Month')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')

# Display the chart
plt.show()
```
